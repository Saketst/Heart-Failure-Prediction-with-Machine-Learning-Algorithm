### *$$Treating~ Outliers$$*

```python
for i in l1:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    df = df[~((df[i]<(Q1-1.5*IQR))|(df[i]>(Q3+1.5*IQR)))]
    plt.title(i)
    sns.boxplot(df[i])
    plt.show()
```

**Using IQR method we have removed the outliers. If in case some outliers are still present according to boxplot, they are to be treated as data points not as outliers.**

### *$$Oversampling$$*

```python
class_count_0, class_count_1 = df.DEATH_EVENT.value_counts()
# Separate class
class_0 = df[df.DEATH_EVENT == 0]
class_1 = df[df.DEATH_EVENT == 1]

print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
    
class_1_over = class_1.sample(class_count_0, replace=True)
df = pd.concat([class_0,class_1_over], axis=0)

print("total values of 1 and 0:",df.DEATH_EVENT.value_counts())

df.DEATH_EVENT.value_counts().plot(kind='bar')
plt.show()
```

# *$$Machine~Learning~Model$$*

```python
from sklearn.model_selection import train_test_split
```

## *$$Separating~Target~Variables~from~other~features$$*

```python
X = df.drop(columns = ["DEATH_EVENT"])
y = df["DEATH_EVENT"]
```

## *$$Separating~Target~Variables~from~other~features:$$*

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30,random_state = 100)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## *$$Feature Scaling$$*

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

___

## *$$Decision~Tree$$*

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
sns.heatmap(confusion_matrix(y_test,predictions), annot = True, fmt = "d")
```
